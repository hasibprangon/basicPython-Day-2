# Conditional Statement

```python
raining = True

if not raining :
    print("Play Football")

else:
    print("Play video games") 
```

## Nested If else

```python
raining = input()
muddy = input()

if raining: 
    if muddy:
        print("Play football & be safe")
    else: 
        print("play football")

else:
    print("Play video games")
```

## Else If

```python
money = int(input())

if money >= 500 :
    print("Going with friends")
elif money <= 400 :
    print("Order food")
elif money <= 300 :
    print("Cook food")
else :
    print("Save the money")
```

## continue and break

```python
sum = 0

for i in range(1, 11):
    # print(i, "is processing....")
    
    if(i % 2 == 0):
        continue
    sum += i
    # print(i, "is added to the sum")
    

print(sum)
```
