## Even, Odd, Negetive, and Positive

```python
inp = int(input())
numbers = []

for i in range(inp):
    num = int(input())
    numbers.append(num)

negetive = 0
positive = 0
odd = 0
even = 0

for i in range(inp):
    x = numbers[i]

    # negetive and positive
    if x > 0:
        positive += 1
    elif x < 0:
        negetive += 1

    # odd even
    if x % 2 != 0:
        odd += 1
    else:
        even += 1

print(negetive, "Negetive")
print(positive, "Positive")
print(odd, "Odd")
print(even, "Even")
```
